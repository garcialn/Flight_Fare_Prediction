###### Here we notice that our dataset is made out of string objects, but the Price column, wich is our Target feature.

###### With that, we have to transform some of our data so that our model can understand and run the code.

```python
# Lets start with the date feature

data_train['Journey_day'] = pd.to_datetime(data_train.Date_of_Journey, format = '%d/%m/%Y').dt.day
data_train['Journey_month'] = pd.to_datetime(data_train['Date_of_Journey'], format = '%d/%m/%Y').dt.month

data_train.drop(['Date_of_Journey'], axis = 1, inplace = True)
```

```python
# Now lets change the departure time feature

data_train['Dep_hour'] = pd.to_datetime(data_train['Dep_Time']).dt.hour
data_train['Dep_min'] = pd.to_datetime(data_train['Dep_Time']).dt.minute

data_train.drop(['Dep_Time'], axis = 1, inplace = True)

# And the arrival time

data_train['Arrival_hour'] = pd.to_datetime(data_train.Arrival_Time).dt.hour
data_train['Arrival_min'] = pd.to_datetime(data_train.Arrival_Time).dt.minute

data_train.drop(['Arrival_Time'], axis = 1, inplace = True)
```

```python
data_train
```

```python
# Now we're gonna transform the Duration column into a int variable

duration = list(data_train['Duration']) # Turning the column series into a list

for i in range(len(duration)):
    if len(duration[i].split()) != 2:   # Cheking if every duration value is make out of hours and minutes or days
        if 'h' in duration[i]:
            duration[i] = duration[i].strip() + ' 0m'  # Adds 0 minute to the duration value
        else:
            duration[i] = '0h ' + duration[i]          # Adds 0 hours to the duration value

duration_hours = []  # Creating a list for the hours
duration_mins = []   # Creating a list for the minutes

for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = 'h')[0]))             # We'll be adding the hours (int) into the list
    duration_mins.append(int(duration[i].split(sep = 'm')[0].split()[-1]))  # Adding the minutes (int) into the the list after the hours
    
# Adding the hours and minutes list into our dataframe    
data_train['Duration_hours'] = duration_hours
data_train['Duration_mins'] = duration_mins
```
